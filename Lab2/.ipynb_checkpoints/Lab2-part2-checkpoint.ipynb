Categorize: Create a new list/array/dataframe containing type corresponding to
sales, name this as “sales type”. If sales is >=15, set value of sales type to
“high”, if sales is <15 and >=5, set value of sales type to “medium”, else set
sales to “low”. (If you chose data of your choice, create a new category with
respect to your data and follow this in further steps as well.)

```python
salestype = pd.DataFrame(columns=["Type"])
salestype.head()
```

```python
def convert(x):
    if(x>=15):
        return "High"
    
    elif(x<15 and x>=5):
        return "Medium"
    
    else:
        return "Low"
```

```python
salestype['Type'] = data.Sales.apply(convert)
```

```python
salestype.head(10)
```

```python
#Concatenation
data = pd.concat([data,salestype],axis = 1)
data.head(10)
```
