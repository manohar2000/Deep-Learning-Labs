## Creating a dataframe (pandas) 

5 entries, created randomly. List of features would
be “student name”,”date of birth”, “marks in maths”,

```python
# Create this dataframe from  lists of lists.

import pandas as pd  
  
data = [['Chintu', 18, 80], ['Pintu', 19, 79], ['Tintu', 20 , 90]]  
   
df = pd.DataFrame(data, columns = ['Name', 'Age' , 'Marks'])  
  
print(df)
```

```python
# Create this dataframe from  dicts of ndarray

data = { "Name" : ["Chintu","Pintu","Tintu"],
         "Age" : [18,19,20],
        "Marks" : [75,80,90]
       }

df = pd.DataFrame(data)  
  
print(df)
```

```python
# Create this dataframe from list of dicts

data = [{'Name': 'Chintu', 'Age': 18, 'Marks': 75}, 
        {'Name': 'Pintu', 'Age': 19, 'Marks': 80},
        {'Name': 'Tintu', 'Age': 20, 'Marks': 90}
       ]  
df = pd.DataFrame(data)
print(df)
```

```python
# Create this dataframe from dict of series


data = {'Chintu': 18 , 
        'Pintu' : 19 ,
        'Tintu' : 20 }
        
  
# create a series 
series = pd.Series(data)
series
```

```python
df = pd.DataFrame(series, columns = ["Age"])
df
```

## Working with series

a. Create two panda series of 100 random numbers  <br />
b. Write a Pandas program to add, subtract, multiply and divide two Pandas Series <br />
c. Store all the 6 series to a dataframe with column names as “number 1” “number 2”,
“add”, “sub”, “mult” and “div”


```python
df1 = pd.DataFrame(np.random.randint(0,100,size=(100, 1)), columns=['Number 1'])
df1
```

```python
df2 = pd.DataFrame(np.random.randint(0,100,size=(100, 1)), columns=['Number 2'])
df2
```

```python
add = pd.DataFrame(np.add(df1,df2).values,columns=["Add"])
add
```

```python
subtract = pd.DataFrame(np.subtract(df1,df2).values,columns=["Subtract"])
add
```

```python
multiply = pd.DataFrame(np.multiply(df1,df2).values,columns=["Mutiply"])
multiply.head(10)
```

```python
div = pd.DataFrame(np.true_divide(df1,df2).values,columns=["Division"])
div.head(10)
```

Store all the 6 series to a dataframe with column names as “number 1” “number 2”, “add”, “sub”, “mult” and “div”

```python
combined = pd.concat([df1, df2,add,subtract,multiply,multiply,div], axis=1)
combined
```
