Fill in missing values: There are a variety of methods which can be used to fill in the missing values (Recently Created). Explore them <br />
a. Fill missing(NaN/NULL) values with 0 <br />
b. Fill missing(NaN/NULL) values with average of column

```python
df["Newspaper"].fillna(0,inplace=True)
df["TV"].fillna(0,inplace=True)
df.head()
```

```python
#df.fillna(axis = 0 , method = df.mean(),inplace = True)
df["Newspaper"].fillna(df["Newspaper"].mean(), inplace = True) 
df["TV"].fillna(df["TV"].mean(),inplace=True)
df.head()
```

```python
y = df.Sales
X = df.drop(["Sales"],axis=1)
```

```python
y.head(10)
```

```python
X.head()
```

Normalization: Each column can be normalized into 0 to 1 using different methods such
as scaling, standardizing etc. This part you have to explore. Note: Normalization should
not be done for the target feature(which is sales here).

```python
numerical_df = X.select_dtypes(include=[np.number])
numerical_df.head()
```

```python
# mean normalization 
normalized_df=(numerical_df-numerical_df.mean())/numerical_df.std()
normalized_df.head()
```

```python
#min-max scaling
normalized_df=(numerical_df-numerical_df.min())/(numerical_df.max()-numerical_df.min())
normalized_df.head()
```

```python
print(normalized_df.shape)
normalized_df.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(normalized_df,y,test_size=0.40)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,predictions))
```
