```
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

stop_words = set(stopwords.words('english')) 
for i in data_positive:
  
  
    
  word_tokens = word_tokenize(i) 
    
  filtered_sentence_positive = [w for w in word_tokens if not w in stop_words] 
    
  filtered_sentence_positive = [] 
    
  for w in word_tokens: 
      if w not in stop_words: 
          filtered_sentence_positive.append(w) 
    
  print(word_tokens) 
  print(filtered_sentence_positive) 
```

```
for i in data_negative: 
    
  word_tokens = word_tokenize(i) 
    
  filtered_sentence_positive = [w for w in word_tokens if not w in stop_words] 
    
  filtered_sentence_negative = [] 
    
  for w in word_tokens: 
      if w not in stop_words: 
          filtered_sentence_negative.append(w) 
    
  print(word_tokens) 
  print(filtered_sentence_negative) 
```

```
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
   
ps = PorterStemmer() 
  
for i in filtered_sentence_negative:
# choose some words to be stemmed   
  for w in i: 
      print(w, " : ", ps.stem(w)) 
```
